```python
import numpy as np
from si.io.csv_file import read_csv
from si.data.dataset import Dataset

# === Load the iris dataset ===
iris_data = read_csv("../datasets/iris/iris.csv", label="class")

print(f"Initial dataset dimensions: {iris_data.shape()}")

# ============================================================
#                   DROPNA DEMONSTRATION
# ============================================================

# Create a new dataset copy with missing values injected
iris_drop = Dataset(
    X=iris_data.X.copy(),
    y=iris_data.y.copy(),
    features=iris_data.features,
    label=iris_data.label
)

# Introduce missing values
iris_drop.X[[0, 10], [1, 2]] = np.nan

print("\n[DROPNA] Samples before dropna:", iris_drop.shape()[0])
iris_drop.dropna()
print("[DROPNA] Samples after dropna:", iris_drop.shape()[0])

# ============================================================
#                   FILLNA DEMONSTRATION
# ============================================================

iris_fill = Dataset( X=iris_data.X.copy(), y=iris_data.y.copy(), features=iris_data.features, label=iris_data.label)

# Insert missing data
iris_fill.X[5, 0] = np.nan
iris_fill.X[15, 3] = np.nan

nan_before = np.isnan(iris_fill.X).sum()
print(f"\n[FILLNA] Missing values before fillna: {nan_before}")

iris_fill.fillna("mean")

nan_after = np.isnan(iris_fill.X).sum()
print(f"[FILLNA] Missing values after fillna:  {nan_after}")

# ============================================================
#               REMOVE BY INDEX DEMONSTRATION
# ============================================================

iris_removed = Dataset(
    X=iris_data.X.copy(),
    y=iris_data.y.copy(),
    features=iris_data.features,
    label=iris_data.label
)

print(f"\n[REMOVE] Samples before removing index: {iris_removed.shape()[0]}")
iris_removed.remove_by_index(0)
print(f"[REMOVE] Samples after removing index:  {iris_removed.shape()[0]}")
```
