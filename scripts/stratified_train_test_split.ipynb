```python
from si.io.csv_file import read_csv
from si.model_selection.split import stratified_train_test_split
import numpy as np

# Load the dataset directly as a Dataset object
iris = read_csv('../datasets/iris/iris.csv', sep=',', features=True, label=True)

# Apply stratified split
train_dataset, test_dataset = stratified_train_test_split(dataset=iris,test_size=0.2,random_state=42)
print("Train dataset class distribution:")
print(np.unique(train_dataset.y, return_counts=True))

print("Test dataset class distribution:")
print(np.unique(test_dataset.y, return_counts=True))
```
