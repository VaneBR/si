1.2) Select the penultimate independent variable.

1.3) Select the last 10 samples from the iris dataset. 
What is the mean of the last 10 s?

```python
last_10_samples = iris_dataset.X[-10:, :]
mean_last_10 = np.mean(last_10_samples, axis=0)
print(f"Mean of the last 10 samples for each feature:")
for i, feature_name in enumerate(iris_dataset.features):
    print(f"  {feature_name}: {mean_last_10[i]:.4f}")
```

1.4) Select all samples from the dataset with values less than or equal to 6 for all independent variables/features. How many samples do you obtain?

```python
mask = np.all(iris_dataset.X <= 6, axis=1)
samples_le_6 = iris_dataset.X[mask]
print(f"Number of samples obtained: {samples_le_6.shape[0]}")
```

1.5) Select all samples with a class/label different from 'Iris-setosa'. How many samples do you obtain?
