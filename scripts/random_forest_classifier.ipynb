```python
# Jupyter Notebook para testar RandomForestClassifier com iris.csv
# Exerc√≠cio 9.2

import numpy as np
import pandas as pd
from si.data.dataset import Dataset
from si.io.csv_file import read_csv
from si.models.random_forest_classifier import RandomForestClassifier
from si.model_selection.split import train_test_split
from si.metrics.accuracy import accuracy

data = read_csv('../datasets/iris/iris.csv', sep=',', features=True, label=True)
train, test = train_test_split(data, test_size=0.2, random_state=42)

# Create the RandomForestClassifier model
rf = RandomForestClassifier(n_estimators=20, max_depth=5, seed=42)

# Train the model
rf.fit(train)

# Evaluate on test set
score = rf.score(test)
print(f"Test accuracy (%): {score * 100:.2f}%")
```
