```python
from si.io.csv_file import read_csv
from si.model_selection.split import train_test_split
from si.ensemble.stacking_classifier import StackingClassifier
from si.models.knn_classifier import KNNClassifier
from si.models.logistic_regression import LogisticRegression
from si.models.decision_tree_classifier import DecisionTreeClassifier
```

```python
#Carregar breat-bin.csv
data=read_csv ('../datasets/breast_bin/breast-bin.csv', sep=',', features=True, label=True)
```

```python
#2. Split
train,test=train_test_split(data, test_size=0.3, random_state=42)
```

```python
#3-5 Criar modelos bases
base_models=[
    KNNClassifier(k=3),
    LogisticRegression(l2_penalty=0.1, alpha=0.1, max_iter=1000),
    DecisionTreeClassifier(max_depth=5)
]
```

```python
#6. Modelo final
final_model = KNNClassifier(k=5)
```

```python
#7. Criar stacking
stacking = StackingClassifier(
    models=base_models, 
    final_model=final_model
)
```

```python
#8. Treinar e avaliar
stacking.fit(train)
print(f"Score: {stacking.score(test): .4f}")
```
